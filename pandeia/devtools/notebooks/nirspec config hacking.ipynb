### Perform an example fit for one grating and aperture to show that a quadratic fit is sufficient.

```python
wavecal = fits.open("/Users/pickering/STScI/smite/pandeia_data/jwst/nirspec/wavecal/g395h_sca491_1dwave.fits")
plt.figure(figsize=(6, 6))
w = wavecal[1].data['A400'][0]
wd = np.diff(w)
wp = np.arange(w.size)
fit = np.polyfit(wp, w, 2)
fit_w = np.poly1d(fit)
#plt.plot(wp, w - fit_w(wp))
#plt.plot(np.arange(wd.size), wd)
plt.plot(w)
fit
```

```python
"""
Here we loop through gratings and determine the mean wavelength solution averaged across apertures.
We average the slope and quadratic terms to get one solution per grating that can be shifted in 
zeropoint/intercept as needed.

The results for B200 and A1600 look a little squirrely, though.  A1600 will have resolution effects due to it large
size while B200 is shifted so far it doesn't work for several gratings.
"""
gratings = ["g140m", "g140h", "g235m", "g235h", "g395m", "g395h"]
apertures = ["B200", "A1600", "A400", "A200_2", "A200_1"]
slopes = {}
intercepts = {}
quads = {}
for g in gratings:
    slopes[g] = 0.0
    intercepts[g] = 0.0
    quads[g] = 0.0
    n = 0
    chip1_file = "/Users/pickering/STScI/smite/pandeia_data/jwst/nirspec/wavecal/%s_sca491_1dwave.fits" % g
    chip2_file = "/Users/pickering/STScI/smite/pandeia_data/jwst/nirspec/wavecal/%s_sca492_1dwave.fits" % g
    wavecal_1 = fits.open(chip1_file)
    wavecal_2 = fits.open(chip2_file)
    print "%s Chip #1:" % g
    dataset = wavecal_1[1].data
    for a in apertures:
        w = dataset[a][0][:-1]
        if w[-1] - w[0] == 0.0:
            print "%8s: N/A" % a
            continue
        wd = np.diff(dataset[a][0])
        wp = np.arange(w.size)
        fit = np.polyfit(wp, w, 2)
        if a not in ['B200', 'A1600']:
            quads[g] += fit[0]
            slopes[g] += fit[1]
            intercepts[g] += fit[2]
            n += 1
        print "%8s: slope=%.5e  intercept=%.8f  quadratic=%.5e" % (a, fit[1], fit[2], fit[0])
    print "%s Chip #2:" % g
    dataset = wavecal_2[1].data
    for a in apertures:
        w = dataset[a][0][:-1]
        if w[-1] - w[0] == 0.0:
            print "%8s: N/A" % a
            continue
        wd = np.diff(dataset[a][0])
        wp = np.arange(w.size)
        fit = np.polyfit(wp, w, 2)
        if a not in ['B200', 'A1600']:
            quads[g] += fit[0]
            slopes[g] += fit[1]
            intercepts[g] += fit[2]
            n += 1
        print "%8s: slope=%.5e  intercept=%.8f  quadratic=%.5e" % (a, fit[1], fit[2], fit[0])
    wavecal_1.close()
    wavecal_2.close()
    quads[g] /= n
    slopes[g] /= n
    intercepts[g] /= n
```

```python
"""
prototype how to read the resolution data out of the csv files
"""
from astropy.io import ascii
r_file = "/Users/pickering/STScI/smite/pandeia_data/jwst/nirspec/dispersion/prism_r.csv"
data = ascii.read(r_file, data_start=4)
w = data['col1'].data
r = data['col2'].data
plt.figure(figsize=(6, 6))
plt.plot(w, r)
```

```python
pix = np.arange(30000)
disp_ref = {}
for g in slopes:
    """
    We get R out of the CSV data and fit a line to it.  For gratings R is defined to be linear because
    it's set by the grating spacing.  This isn't true for transmissive elements where the dispersion
    comes from differential index of refraction which varies with wavelength.
    """
    r_file = "/Users/pickering/STScI/smite/pandeia_data/jwst/nirspec/resolving_power/%s_r.csv" % g
    data = ascii.read(r_file, data_start=4)
    w_r = data['col1'].data
    r = data['col2'].data
    r_fit = np.polyfit(w_r, r, 1)
    r_func = np.poly1d(r_fit)
    # get the fit parameters for the grating. shift the intercept to 0.5 um and calc from there.
    s = slopes[g]
    i = 0.5
    q = quads[g]
    func = np.poly1d([q, s, i])
    wave = func(pix)
    # limit 6 um
    wave_subs = np.where(wave <= 6.0)
    w = wave[wave_subs]
    dlds = np.diff(w)
    r = r_func(w)
    disp_ref[g] = np.arange(dlds.size), w[:-1], dlds, r
    
    # write the info out to FITS tables
    outfile = "%s_disp.fits" % g
    w_col = fits.Column(name='WAVELENGTH', format='1D', unit='UM', disp="F10.1", array=w[:-1])
    disp_col = fits.Column(name='DLDS', format='1D', unit='UM', disp="G12.5", array=dlds)
    r_col = fits.Column(name='R', format='1D', unit='TRANSMISSION', disp="G12.5", array=r[:-1])
    cols = fits.ColDefs([w_col, disp_col, r_col])
    tb_hdu = fits.new_table(cols)
    hdr = fits.Header()
    pri_hdu = fits.PrimaryHDU(header=hdr)
    out_hdu = fits.HDUList([pri_hdu, tb_hdu])
    out_hdu.writeto(outfile)
```

```python
"""
read in MSA shutter information provided by @nirspec
"""
from pandeia.engine.io_utils import write_json, read_json
f = open("/Users/pickering/STScI/smite/pandeia_data/jwst/nirspec/shutters.txt")
lines = f.readlines()
f.close()
config = {}
for l in lines[1:]:
    shutter, grating, filt, w_min, gap_s, gap_e, w_max = l.split()
    w_min, w_max = float(w_min), float(w_max)
    if gap_s == "null":
        gap_s = None
    else:
        gap_s = float(gap_s)
    if gap_e == "null":
        gap_e = None
    else:
        gap_e = float(gap_e)
    if shutter not in config:
        config[shutter] = {}
        config[shutter]['display_string'] = shutter.upper()
    config[shutter][grating] = {}
    config[shutter][grating]['wave_min'] = w_min
    config[shutter][grating]['wave_max'] = w_max
    config[shutter][grating]['gap_start'] = gap_s
    config[shutter][grating]['gap_end'] = gap_e
write_json(config, "shutter_gap.json", sort_keys=True)
```
