```python
import sys
import json
import copy
import numpy as np
import matplotlib.pyplot as plt
from pandeia.engine.perform_calculation import perform_calculation
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
# set up default line configuration
line_strength = 1.0e-12
line = {
    'id': 1,
    'name': 'H Alpha',
    'center': 1.5,
    'width': 300.0,
    'strength': line_strength,
    'profile': 'gaussian',
    'emission_or_absorption': 'emission',
}

# set up parameters to vary for each source
sx = [0, 0.1, 0.15, 0.13, 0.1, 0.05, -0.02, -0.1, -0.14, -0.12, -0.05, -0.0, 0.07, 0.12]
sy = [-0.1, -0.14, -0.12, -0.05, -0.0, 0.07, 0.12, 0, 0.1, 0.15, 0.13, 0.1, 0.05, -0.02]
sf = [0.1, 0.05, 0.01, 0.2, 0.4, 0.2, 0.22, 0.08, 0.12, 0.2, 0.22, 0.34, 0.45, 0.09]

# set up default source configuration
source1 = {
    'id': 1,
    'target': True,
    'position': {
        'ang_unit': 'arcsec',
        'xoff': 0.0,
        'yoff': 0.0,
    },
    'shape': {
        'pa': -45.0,
        'major': 0.6,
        'minor': 0.2
    },
    # flat continuum spectrum normalized to 0.1 mJy
    'spectrum': {
        'w_unit': 'micron',
        'f_unit': 'millijansky',
        'type': 'flat_fnu',
        'flux_norm': 0.1,
        'ref_wave': 2.5,
        'wmin': 0.5,
        'wmax': 2.6,
        'temp': 1000.0,
        'lines': [line]
    }
}

# now assemble the scene
scene = [source1]
for i in range(len(sx)):
    source = copy.deepcopy(source1)
    source['position']['xoff'] = sx[i] * 5
    source['position']['yoff'] = sy[i] * 5
    source['spectrum']['lines'][0]['strength'] = 1.0e-12 * sf[i]
    source['spectrum']['flux_norm'] = 0.001

    source['shape']['major'] = 0.05
    source['shape']['minor'] = 0.05
    scene.append(source)

# set up the rest of the calculation
background = 'medium'
obsmode = {
    'instrument': 'niriss',
    'mode': 'soss',
    'filter': None,
    'aperture': 'imager',
    'disperser': 'gr700xd_1'
}
exp_config = {
    'subarray': 'full',
    'readmode': 'nisrapid',
    'ngroup': 10,
    'nint': 1,
    'nexp': 10
}
strategy = {
    'method': 'specapphot',
    'aperture_size': 0.07,
    'sky_annulus': [0.2, 0.4],
    'target': [0.0, 0.0],
    'dithers': [{'x': 0.0, 'y': 0.0}]
}

calc_input = {
    'scene': scene,
    'background': background,
    'configuration': {'instrument': obsmode,
                      'detector': exp_config},
    'strategy': strategy
}
```

```python
report = perform_calculation(calc_input, dict_report=False)
report_dict = report.as_dict()
report_fits = report.as_fits()
```

```python
plt.figure(figsize=(10,10))
t = report_dict['transform']
if report_dict['information']['calc_type'] == 'image':
    xmin = t['x_min'] - 0.5*t['x_step']
    xmax = t['x_max'] - 0.5*t['x_step']
    aspect = 1.0
else:
    xmin = t['wave_min']
    xmax = t['wave_max']
    aspect = 0.75
ymin = t['y_min'] - 0.5*t['y_step']
ymax = t['y_max'] - 0.5*t['y_step']
extent = [xmin, xmax, ymin, ymax]
implot = plt.imshow(report_dict['2d']['detector'], interpolation='nearest', extent=extent, aspect=aspect*(xmax-xmin)/(ymax-ymin))

strategy = calc_input['strategy']
circles = []
circles.append(plt.Circle((0.0, 0.0), radius=strategy['aperture_size'], edgecolor='white', facecolor='none'))
circles.append(plt.Circle((0.0, 0.0), radius=strategy['sky_annulus'][0], edgecolor='red', facecolor='none'))
circles.append(plt.Circle((0.0, 0.0), radius=strategy['sky_annulus'][1], edgecolor='red', facecolor='none'))
for c in circles:
    implot.axes.add_artist(c)
plt.colorbar(orientation='horizontal')
```
