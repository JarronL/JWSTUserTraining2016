```python
#sp = psyn.FileSpectrum("/Users/pickering/STScI/smite/pandeia_data/sed/galaxies/ngc_7771_spec.fits")
sp = psyn.Icat("phoenix", 3500.0, -4.0, 1.0)
sp.convert('microns')
sp.convert('mjy')
wave, flux = sp.wave, sp.flux
subs = np.where((wave > 0.7) & (wave < 10.0))
wave = wave[subs]
flux = flux[subs]
```

```python
#sp = psyn.FileSpectrum("/Users/pickering/STScI/smite/pandeia_data/sed/galaxies/ngc_7771_spec.fits")
sp_l = psyn.Icat("phoenix", 3500.0, -3.5, 4.0)
sp_l.convert('microns')
sp_l.convert('mjy')
wave_l, flux_l = sp_l.wave, sp_l.flux
subs = np.where((wave_l > 0.7) & (wave_l < 10.0))
wave_l = wave_l[subs]
flux_l = flux_l[subs]
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.gca()
ax.set_yscale('log')

plt.plot(wave, flux)
plt.plot(wave_l, flux_l)
plt.ylabel(r"F_nu (mJy)")
plt.xlabel(r"wavelength (um)")
```

```python
from astropy.table import Table
t = Table.read("/Users/pickering/STScI/galaxy_sed_catalog/summary.csv", format="ascii.csv")
```

```python
cols = [c.lower() for c in t.colnames]
ucols = t.colnames
zip(cols[1:], ucols[1:])
```

```python
config = {}
for i in range(len(t)):
    key = t['Name'][i]
    filename = key.replace(" ", "_").lower() + "_spec.fits"
    config[key] = {}
    config[key]['filename'] = filename
    for c in ucols[1:]:
        config[t['Name'][i]][c] = t[c][i]
```
