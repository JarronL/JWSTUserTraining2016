```python
# set up the various implementations for calculating the covariance matrix
import scipy.sparse as sparse
def fill_array(n, val):
    array = np.empty(n, np.dtype('i2'))
    array.fill(val)
    return array

def create_covariance_array(signal, noise, subscripts):
    diagonal = noise.var_pix[subscripts].ravel()
    nn = diagonal.shape[0]
    
    ncorr = int((signal.read_noise_correlation_matrix.shape[0] - 1) / 2)
    c_ij_dense = np.diagflat(diagonal).astype(np.float64)
    for i in range(nn):
        correlation_subscripts = (subscripts[0] - subscripts[0][i] + ncorr,
                                  subscripts[1] - subscripts[1][i] + ncorr)
        c_ij_dense[i, :] = signal.read_noise_correlation_matrix[correlation_subscripts] * diagonal[i]
    return c_ij_dense

def create_covariance_matrix(signal, noise, subscripts):
    diagonal = noise.var_pix[subscripts].ravel()
    nn = diagonal.shape[0]
    
    ncorr = int((signal.read_noise_correlation_matrix.shape[0] - 1) / 2)
    c_ij_dense = np.diagflat(diagonal)
    for i in range(nn):
        correlation_subscripts = (subscripts[0] - subscripts[0][i] + ncorr,
                                  subscripts[1] - subscripts[1][i] + ncorr)
        c_ij_dense[i, :] = signal.read_noise_correlation_matrix[correlation_subscripts] * diagonal[i]
    c_ij = np.matrix(c_ij_dense)
    return c_ij

def create_covariance_matrix_sparse(signal, noise, subscripts):
    diagonal = noise.var_pix[subscripts].ravel()
    nn = diagonal.shape[0]
    
    ncorr = int((signal.read_noise_correlation_matrix.shape[0] - 1) / 2)
    c_ij_dense = np.zeros((nn, nn))
    for i in range(nn):
        correlation_subscripts = (subscripts[0] - subscripts[0][i] + ncorr,
                                  subscripts[1] - subscripts[1][i] + ncorr)
        c_ij_dense[i, :] = signal.read_noise_correlation_matrix[correlation_subscripts] * diagonal[i]
    c_ij = sparse.csr_matrix(c_ij_dense)
    c_ij.setdiag(diagonal)
    return c_ij
```

```python
# this cell implements Strategy._create_weight_matrix()
n = noise_list[0]
s = signal_list[0]
aperture = 0.3
annulus = [0.5, 0.6]
ny, nw = n.var_pix.shape
ysamp = s.grid.ysamp

weight_matrix = np.matrix(np.zeros((ny, nw)))
aper_range = (np.ceil((ny - 1) / 2. - aperture / ysamp),
              np.floor((ny - 1) / 2. + aperture / ysamp))
sky_range1 = (np.ceil((ny - 1) / 2. + annulus[0] / ysamp),
              np.floor((ny - 1) / 2. + annulus[1] / ysamp))
sky_range2 = (np.ceil((ny - 1) / 2. - annulus[1] / ysamp),
              np.floor((ny - 1) / 2. - annulus[0] / ysamp))

weight_matrix[aper_range[0]:aper_range[1] + 1, :] = 1.

weight_matrix[sky_range1[0]:sky_range1[1] + 1, :] = \
    -(aper_range[1] - aper_range[0] + 1) / ((sky_range1[1] - sky_range1[0]) + (sky_range2[1] - sky_range2[0]) + 2)

weight_matrix[sky_range2[0]:sky_range2[1] + 1, :] = \
    -(aper_range[1] - aper_range[0] + 1) / ((sky_range1[1] - sky_range1[0]) + (sky_range2[1] - sky_range2[0]) + 2)

colsubs = np.arange(ny, dtype=np.dtype('i4'))

subscripts = [(colsubs, fill_array(ny, i)) for i in np.arange(nw)]
a_ij = weight_matrix
```

```python
%%timeit
# this cell implements Strategy._error_sum() using sparse matrices, but with the matrix initialized outside of the loop
# and then scaled at each step
nproducts = len(subscripts)
flux_products = np.empty(nproducts)
flux_plus_bg_products = np.empty(nproducts)
sigma_products = np.empty(nproducts)
flux_tots = np.empty(nproducts)
flux_plus_bg_tots = np.empty(nproducts)

init_scale = n.var_pix[subscripts[0]].ravel()
c_ij_init = create_covariance_matrix_sparse(s, n, subscripts=subscripts[0])
for product_subscript, i in zip(subscripts, np.arange(nproducts)):
    a_ij_raveled = a_ij[product_subscript]
    
    # divide current row by first row, transform into diagonal matrix, and multiply by transpose of c_ij so
    # multiplication is done row-wise
    scaled = c_ij_init.transpose() * np.matrix(np.diagflat(n.var_pix[product_subscript].ravel()/init_scale))
    c_ij = scaled.transpose()
    
    var_product = a_ij_raveled * c_ij * a_ij_raveled.transpose()
    flux_product = a_ij_raveled.dot(np.matrix(s.rate[product_subscript]).transpose())
    flux_plus_bg_product = a_ij_raveled.dot(np.matrix(s.rate_plus_bg[product_subscript]).transpose())
    flux_tot = np.sum(s.rate[product_subscript])
    flux_plus_bg_tot = np.sum(s.rate_plus_bg[product_subscript])

    sigma_products[i] = np.sqrt(var_product.item())
    flux_products[i] = flux_product.item()
    flux_plus_bg_products[i] = flux_plus_bg_product.item()
    flux_tots[i] = flux_tot.item()
    flux_plus_bg_tots[i] = flux_plus_bg_tot.item()
print sigma_products
```

```python
%%timeit
# this cell implements Strategy._error_sum() using normal matrices
nproducts = len(subscripts)
flux_products = np.empty(nproducts)
flux_plus_bg_products = np.empty(nproducts)
sigma_products = np.empty(nproducts)
flux_tots = np.empty(nproducts)
flux_plus_bg_tots = np.empty(nproducts)

init_scale = n.var_pix[subscripts[0]].ravel()
c_ij_init = create_covariance_matrix(s, n, subscripts=subscripts[0])
for product_subscript, i in zip(subscripts, np.arange(nproducts)):
    a_ij_raveled = a_ij[product_subscript]
    
    # divide current row by first row, transform into diagonal matrix, and multiply by transpose of c_ij so
    # multiplication is done row-wise
    scaled = c_ij_init.transpose() * np.matrix(np.diagflat(n.var_pix[product_subscript].ravel()/init_scale))
    c_ij = scaled.transpose()
    
    var_product = a_ij_raveled * c_ij * a_ij_raveled.transpose()
    flux_product = a_ij_raveled.dot(np.matrix(s.rate[product_subscript]).transpose())
    flux_plus_bg_product = a_ij_raveled.dot(np.matrix(s.rate_plus_bg[product_subscript]).transpose())
    flux_tot = np.sum(s.rate[product_subscript])
    flux_plus_bg_tot = np.sum(s.rate_plus_bg[product_subscript])

    sigma_products[i] = np.sqrt(var_product.item())
    flux_products[i] = flux_product.item()
    flux_plus_bg_products[i] = flux_plus_bg_product.item()
    flux_tots[i] = flux_tot.item()
    flux_plus_bg_tots[i] = flux_plus_bg_tot.item()
print sigma_products
```

```python
%%timeit
# this cell implements Strategy._error_sum() using numpy.ndarrays
nproducts = len(subscripts)
flux_products = np.empty(nproducts)
flux_plus_bg_products = np.empty(nproducts)
sigma_products = np.empty(nproducts)
flux_tots = np.empty(nproducts)
flux_plus_bg_tots = np.empty(nproducts)

init_var = n.var_pix[subscripts[0]].ravel()
c_ij_init = create_covariance_array(s, n, subscripts=subscripts[0])
c_ij_norm = c_ij_init / init_var.reshape(len(init_var), 1)

for product_subscript, i in zip(subscripts, np.arange(nproducts)):
    a_ij_raveled = a_ij[product_subscript]
    
    diagonal = n.var_pix[product_subscript].ravel()
    c_ij = np.matrix(c_ij_norm * diagonal.reshape(len(diagonal), 1))
    
    var_product = a_ij_raveled * c_ij * a_ij_raveled.transpose()
    flux_product = a_ij_raveled.dot(np.matrix(s.rate[product_subscript]).transpose())
    flux_plus_bg_product = a_ij_raveled.dot(np.matrix(s.rate_plus_bg[product_subscript]).transpose())
    flux_tot = np.sum(s.rate[product_subscript])
    flux_plus_bg_tot = np.sum(s.rate_plus_bg[product_subscript])

    sigma_products[i] = np.sqrt(var_product.item())
    flux_products[i] = flux_product.item()
    flux_plus_bg_products[i] = flux_plus_bg_product.item()
    flux_tots[i] = flux_tot.item()
    flux_plus_bg_tots[i] = flux_plus_bg_tot.item()
print sigma_products
```

```python
%%timeit
# this cell implements Strategy._error_sum() using the original implementation
nproducts = len(subscripts)
flux_products = np.empty(nproducts)
flux_plus_bg_products = np.empty(nproducts)
sigma_products = np.empty(nproducts)
flux_tots = np.empty(nproducts)
flux_plus_bg_tots = np.empty(nproducts)

for product_subscript, i in zip(subscripts, np.arange(nproducts)):
    a_ij_raveled = a_ij[product_subscript]
    
    c_ij = create_covariance_matrix_sparse(s, n, subscripts=product_subscript)
    
    var_product = a_ij_raveled * c_ij * a_ij_raveled.transpose()
    flux_product = a_ij_raveled.dot(np.matrix(s.rate[product_subscript]).transpose())
    flux_plus_bg_product = a_ij_raveled.dot(np.matrix(s.rate_plus_bg[product_subscript]).transpose())
    flux_tot = np.sum(s.rate[product_subscript])
    flux_plus_bg_tot = np.sum(s.rate_plus_bg[product_subscript])

    sigma_products[i] = np.sqrt(var_product.item())
    flux_products[i] = flux_product.item()
    flux_plus_bg_products[i] = flux_plus_bg_product.item()
    flux_tots[i] = flux_tot.item()
    flux_plus_bg_tots[i] = flux_plus_bg_tot.item()
print sigma_products
```
